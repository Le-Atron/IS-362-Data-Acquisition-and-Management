```python

# first we need to import the movies CSV file.

movies = pd.read_csv('movies.csv', index_col=0)

movies.head()
```

```python
# movies DataFrame

moviesdf = pd.DataFrame(movies)

moviesdf
```

```python
# Average rating by movie

moviesdf.mean()
```

```python
# Next I want to add a column that will display the average rating for each movie by each person. 

moviesdf['Average Rating'] = moviesdf.iloc[:, 1:7].mean(axis=1)


moviesdf
```

```python
# Normalizing the Data

movies_norm = (movies - movies.min()) / (movies.max() - movies.min())

movies_norm
```

```python
movies_norm.mean()
```
