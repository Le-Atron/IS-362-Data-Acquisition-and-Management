```python
# first we need to import the data from the UC Irvine data Repository.

autompg = pd.read_fwf('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', header=None,)

autompg.columns =['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration','model year', 'origin', 'car name' ]

autompg
```

```python
# Replacing ? with nana values

autompg = autompg.replace('?', value=np.nan,)

autompg
```

```python
# Replacing numbers with corresponding region


autompg["origin"].replace({1:'USA', 2:'Asia', 3:'Europe'}, inplace=True)

autompg
```

```python
# Setting horsepower as a float

autompg['horsepower'] = autompg['horsepower'].astype('float')
```

```python
# checking data types

autompg.dtypes
```

```python
# creating a bar chart that shows the distribution for cylinders

sns.countplot(data=autompg, x ='cylinders', hue='origin', )
```

```python
# creating a scatterplot that shows the relationship between horsepower and weight


sns.scatterplot(data=autompg, x='horsepower', y='weight')
```

```python
# lets find out the engine displacment vs the mpg rating with seaborn
# as we can see the hiher the mpg, the lower displacemtt the engine is, and vice versa. 

sns.scatterplot(data=autompg, x ='displacement', y='mpg' )
```

```python
# We can also use MatplotLib to create the same plot graph. 

plt.scatter(data=autompg, x ='displacement', y='mpg' )
plt.title('MPG VS Displacement')
plt.xlabel('Displacement')
plt.ylabel('MPG')
```
