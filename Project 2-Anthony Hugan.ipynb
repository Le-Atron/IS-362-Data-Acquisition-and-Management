```python
# 1 Cases

# first we need to import the cases CSV file i made from messy data 

cases = pd.read_csv('cases.csv')

cases.head()
```

The issue I want to solve with this data set is separating the Rate column into two different columns

```python
new = cases["Rate"].str.split("/", n = 1, expand = True)

cases["Cases"]= new[0]

cases["Population"]= new[1]

cases.drop(columns =["Rate"], inplace = True) 

cases
```

```python
cases
```

```python
# My first calculation gave an error, so I had to convert these columns to Integers to find rate

cases['Cases'] = cases['Cases'].astype('int')

cases['Population'] = cases['Population'].astype('int')
```

```python
# This will add a new column, and calculate the rate of cases. 


cases['Rate'] = cases['Cases']/cases['Population']*100

cases
```

```python
# 2 Students 

# first we need to import the students CSV file 

students = pd.read_csv('students.csv')

students.head()
```

The first thing I want to do is separate the Sex and Age column into two separate ones. 

```python
new2 = students["Sex and Age"].str.split("_", n = 1, expand = True)

students["Sex"]= new2[0]

students["Age"]= new2[1]

students.drop(columns =["Sex and Age"], inplace = True) 

students
```

```python
# Next I want to add a column that will display the average grade for each student. 

students['Average Grade'] = students.iloc[:, 5:7].mean(axis=1)

students
```

```python
# 3 teachers 

# first we need to import the students CSV file 

teachers = pd.read_csv('teachers.csv')

teachers.head()
```

```python
# First I want to break up the address column into street city, and state
```

```python
new3 = teachers["Home Address"].str.split(",", n = 2, expand = True)

teachers["Street"]= new3[0]

teachers["City"]= new3[1]

teachers["State"]= new3[2]

teachers.drop(columns =["Home Address"], inplace = True) 

teachers
```

```python
#I also want to fix the name field, and make it two columns. 

new4 = teachers["Name"].str.split(" ", n = 2, expand = True)

teachers["First Name"]= new4[0]

teachers["Last Name"]= new4[1]

teachers.drop(columns =["Name"], inplace = True) 

teachers
```

```python
teachers.dtypes
```

```python
#I wanted to find the mean salary of all teachers but ran into some issues. 
#I had to remove the commas from the Salary column to be able to convert the strings to integers. Once I did this, 
#I was able to use the mean function to find the mean of all Salary’s. 

teachers.replace(',','', regex=True, inplace=True)

teachers
```

```python
teachers['Salary'] = teachers['Salary'].astype('int')
```
